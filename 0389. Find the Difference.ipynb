# 389. Find the Difference

## Description

<br>Given two strings s and t which consist of only lowercase letters.

<br>String t is generated by random shuffling string s and then add one more letter at a random position.

<br>Find the letter that was added in t.

### Example

<br>Input:
<br>s = "abcd"
<br>t = "abcde"

<br>Output:
<br>e

<br>Explanation:
<br>'e' is the letter that was added.

```python
from collections import Counter
class Solution:
    def findTheDifference(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        ret = Counter(t) - Counter(s)
        return list(ret.keys())[0]
```
